```python
entries = toggl.request("https://api.track.toggl.com/api/v9/me/time_entries", {'start_date': '2023-03-01', 'end_date': '2023-03-31'})
```

```python
df = pd.DataFrame(entries)
df.head()
```

```python
df.info()
```

```python
df['start'] = pd.to_datetime(df.start)
df['stop'] = pd.to_datetime(df.stop)
df.head()
```

```python
workspace_id = toggl.getWorkspaces()[0]['id']
projects = toggl.getWorkspaceProjects(workspace_id)
all_clients = toggl.getClients()
clients = [client for client in all_clients if client['wid'] == workspace_id]
client_id_to_name = {client['id']: client['name'] for client in clients}
project_id_to_name = {project['id']: project['name'] for project in projects}
project_name_to_client_id = {project['name']: client_id_to_name[project['cid']] for project in projects}
```

```python
df.drop(columns=['workspace_id', 'task_id', 'billable', 'server_deleted_at', 'uid','user_id','wid','pid','duronly','tag_ids'], inplace=True)
```

```python
df.loc[df['duration'] < 0, 'duration'] = None
```

```python
# map the project ids to the project names
df['project'] = df['project_id'].map(project_id_to_name)
df['client'] = df['project'].map(project_name_to_client_id)
df.drop(columns=['project_id'], inplace=True)
```

```python
# filter to rows whose client is 'Work' or 'Opt'
df_work_opt = df[df['client'].isin(['Work', 'Opt'])]
# group by day
duration_df = df_work_opt[['start', 'duration','client']].groupby([pd.Grouper(key='start', freq='D'), 'client']).sum()
duration_df['dur_mins'] = duration_df['duration'] / 60
duration_df['dur_hrs'] = duration_df['dur_mins'] / 60
duration_df
```

```python
durs = duration_df.drop(columns=['duration', 'dur_mins']).unstack()
durs.columns = durs.columns.droplevel(0)
durs = durs.fillna(0)
```

```python
durs['Total'] = durs['Opt'] + durs['Work']
durs
```

```python
durs.plot()
durs.plot.bar()
```
