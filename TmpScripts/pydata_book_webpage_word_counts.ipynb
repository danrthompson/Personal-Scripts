```python
word_counts = [
    ("1", 5532),
    ("2", 8381),
    ("3", 11518),
    ("4", 13122),
    ("5", 13401),
    ("6", 8648),
    ("7", 12433),
    ("8", 9555),
    ("9", 6464),
    ("10", 10242),
    ("11", 14783),
    ("12", 5683),
    ("13", 11937),
    ("A", 11387),
    ("B", 8737),
]

filter()
```

```python
def print_time(label, mins):
    mins_rounded = round(mins, 2)
    hours = math.floor(mins / 60)
    mins = round(mins % 60)
    print(f"{label}: {hours}:{mins:02d} ({mins_rounded} minutes)")
```

```python
# make into pandas dataframe
df = pd.DataFrame(word_counts, columns=["chapters", "word_counts"])
df
```

```python
df["word_count_totals"] = df["word_counts"].cumsum()
df
```

```python
df["word_count_cum_pct"] = df["word_count_totals"] / df["word_count_totals"].max()
df["word_count_pct"] = df["word_counts"] / df["word_count_totals"].max()
df
```

```python
df.loc[5, "actual_time"] = time_taken_chp_6
df.loc[6, "actual_time"] = time_taken_chp_7
df.loc[7, "actual_time"] = time_taken_chp_8
df
```

```python
df["projected_total_time"] = df["actual_time"]/df["word_count_pct"]
df
```

```python
ptt = df["projected_total_time"].dropna()
wcp = df["word_count_pct"]
ptt, wcp
```

```python
wcpm = pd.DataFrame(wcp)
pttm = pd.DataFrame(ptt).transpose()
```

```python
wcpm.rename(columns={"word_count_pct": "projected_total_time"}, inplace=True)
```

```python
projected_time = wcpm @ pttm
```
